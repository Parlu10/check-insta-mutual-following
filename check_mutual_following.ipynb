```python
import json

# Loads following file
with open("following.json", "r", encoding="utf-8") as f:
    following_data = json.load(f)

# Loads followers file
with open("followers_1.json", "r", encoding="utf-8") as f:
    followers_data = json.load(f)

following = {
    user["title"]
    for user in following_data["relationships_following"]
}

followers = {
    user["string_list_data"][0]["value"]
    for user in followers_data
}

print("Following:", len(following))
print("Followers:", len(followers))

# Account che NON ricambiano il follow
not_following_back = following - followers

print(f"Trovati {len(not_following_back)} account che non ricambiano il follow:\n")

for username in sorted(not_following_back):
    print(username)
```
